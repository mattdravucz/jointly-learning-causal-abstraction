```python
def avg_distance(results):
    dis = 0.0
    n = len(results)
    for i in range(n):
        dis += results[i]['distance'].item()
    return dis/n

def avg_jsd(results):
    jsd = 0.0
    n = len(results)
    for i in range(n):
        jsd += results[i]['jsd'].item()
    return jsd/n

def avg_penalty(results):
    penalty = 0.0
    n = len(results)
    for i in range(n):
        penalty += results[i]['penalty'].item()
    return penalty/n



def sdt_distance(results):
    n = len(results)
    dis = np.zeros((n))
    for i in range(n):
        tmp = results[i]['distance'].item()
        dis[i] = tmp
    return np.std(dis,axis=0)

def sdt_jsd(results):
    n = len(results)
    jsd = np.zeros((n))
    for i in range(n):
        jsd[i] = results[i]['jsd'].item()
    return np.std(jsd,axis=0)

def sdt_penalty(results):
    n = len(results)
    penalty = np.zeros((n))
    for i in range(n):
        penalty[i] = results[i]['penalty'].item()
    return np.std(penalty,axis=0)




def avg_time(results):
    return np.mean(results*10.0,axis=0)
    
def sdt_time(results):
    return np.std(results*10.0,axis=0)
```
